```python
filename='上市ESG-HR.csv'
file = open(filename, mode='r',encoding='utf-8',newline='')
# OPEN 傳出一個 file 物件
```


# 用工具箱裡面 reader 的方法
# Reader Objects-> 
csvreader.__next__()
Return the next row of the reader’s iterable object as a list (if the object was returned from reader()) or a dict (if it is a DictReader instance), parsed according to the current Dialect. Usually you should call this as next(reader).
__next -> iterable 可以 用 for XXX in XXX: 一個一個抓出來

```python
import csv
csv_reader = csv.reader(file)
data_all = list(csv_reader)
for item in data_all:
    if '0.00%' in item[10]:
        print(item)
```
